```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import datetime as dt


#Read CSV
htag= pd.read_csv('Tags/Snow_Emergency_Howe_Tags_2018.csv',encoding='iso-8859-1')
htow= pd.read_csv('Tows/Snow_Emergency_Howe_Tows_2018.csv')
```

```python
#Create DataFrame
htag_df = pd.DataFrame(htag)
htow_df = pd.DataFrame(htow)
```

```python
#Clean Data
#Change column headers and date format
htag_df=htag_df.drop(columns=['X','Y'])
htow_df=htow_df.drop(columns=['X','Y','Tow_Reason'],axis=1)
htow_df.rename(columns={'Snow_Emerg':'snow_emergency_name','Service_Da':'service_date','Address':'address','Latitude': 'latitude','Longitude':'longitude', 'Ward':'ward','Day':'day_id', 'Neighborho':'neighborhood','Tow_Zone':'tow_district'}, inplace=True)
htag_df.rename(columns={'Snow_Em_Na':'snow_emergency_name','Citation_N':'citation','Service_Da':'service_date','Address':'address','Latitude': 'latitude','Longitude':'longitude','Day_ID':'day_id','Ordinance':'ordinance_id','Ordinanc_1':'ordinance_desc','Ward':'ward','Neighborho':'neighborhood','Community':'community','Tow_Zone':'tow_district'},inplace=True)



```

```python
htag_df['service_date']=pd.to_datetime(htag_df['service_date'].astype(str).str[:10])
htow_df['service_date']=pd.to_datetime(htow_df['service_date'].astype(str).str[:10])
```

```python
#Send to SQL
from sqlalchemy import create_engine

import pymysql
pymysql.install_as_MySQLdb()
engine = create_engine('mysql://root:@localhost/snow_emergency_db')
```

```python
#Truncate table before loading
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
session.execute('''TRUNCATE TABLE howe_tags_base ''')
session.commit()
session.close()

Session = sessionmaker(bind=engine)
session = Session()
session.execute('''TRUNCATE TABLE howe_tows_base''')
session.commit()
session.close()

#To SQL Database
htag_df.to_sql(name='howe_tags_base',con=engine,if_exists='append',index=False)
htow_df.to_sql(name='howe_tows_base',con=engine,if_exists='append',index=False)
```
